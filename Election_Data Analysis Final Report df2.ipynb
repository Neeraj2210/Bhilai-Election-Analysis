```python
df2=pd.read_excel("E:/Project/Resolvent Projects/C.G.Election Analysis/Analysis/Previous Election Data.xlsx")
df2.head()
```

```python
# Increase figure size
fig, ax = plt.subplots(figsize=(14, 8))

# Plot the data
ax.plot(df2['Year'], df2['No. of Voters'], marker='o', linestyle='-', color='green', markersize=8, linewidth=3, label='Number of Voters')

# Improve labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Number of Voters')
ax.set_title('Year-wise Number of Voters')

# Customize grid
ax.grid(True, linestyle='--', alpha=0.7)

# Add data labels
for i, txt in enumerate(df2['No. of Voters']):
    ax.annotate(txt, (df2['Year'][i], df2['No. of Voters'][i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8)

# Beautify legend
ax.legend()

# Mark the years on the x-axis properly
ax.set_xticks(df2['Year'])
ax.set_xticklabels(df2['Year'])

# Save the plot
plt.savefig('E:/Project/Resolvent Projects/C.G.Election Analysis/Analysis/yearly_voters_chart.png')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming df2 is your DataFrame

# Calculate vote cast percentage
df2['Vote Percentage'] = (df2['Vote Casted'] / df2['No. of Voters']) * 100

# Increase figure size
fig, ax = plt.subplots(figsize=(14, 6))

# Bar width
bar_width = 0.35

# Set up indices for bar positions
bar_positions_voters = np.arange(len(df2['Year']))
bar_positions_votes_casted = bar_positions_voters + bar_width

# Bar graph for Number of Voters
ax.bar(bar_positions_voters, df2['No. of Voters'], bar_width, label='Number of Voters', color='blue')

# Bar graph for Vote Casted
bars = ax.bar(bar_positions_votes_casted, df2['Vote Casted'], bar_width, label='Vote Casted', color='orange')

# Labeling the bars with Vote Cast Percentage
for bar, vote_cast_percentage in zip(bars, df2['Vote Percentage']):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height + 0.02 * max(df2['No. of Voters']), f'{vote_cast_percentage:.2f}%', ha='center', va='bottom')

# Improve labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Count')
ax.set_title('Year-wise Number of Voters and Vote Casted with Percentage')
ax.set_xticks(bar_positions_voters + bar_width / 2)
ax.set_xticklabels(df2['Year'], rotation=45, ha='right')  # Rotate year labels for better visibility

# Beautify legend
ax.legend()

# Save the plot before showing it
plt.savefig('E:/Project/Resolvent Projects/C.G.Election Analysis/Analysis/yearlWise number of Voters ad vote casted.png')

# Show the plot
plt.show()
```

```python
# Assuming df2 is your DataFrame

# Calculate percentage of Male and Female Voters
df2['Male Percentage'] = (df2['Male Voters'] / df2['No. of Voters']) * 100
df2['Female Percentage'] = (df2['Female Voters'] / df2['No. of Voters']) * 100

# Increase figure size
fig, ax = plt.subplots(figsize=(14, 6))

# Plotting lines for Male Voters and Female Voters with blue and custom pink colors
ax.plot(df2['Year'], df2['Male Voters'], marker='o', linestyle='-', color='blue', label='Male Voters')
ax.plot(df2['Year'], df2['Female Voters'], marker='o', linestyle='-', color='#AA336A', label='Female Voters')

# Improve labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Count')
ax.set_title('Year-wise Male and Female Voters')

# Add data labels for Male Voters
for i, txt in enumerate(df2['Male Voters']):
    ax.annotate(f'{txt} ({df2["Male Percentage"][i]:.2f}%)', (df2['Year'][i], df2['Male Voters'][i]), textcoords="offset points", xytext=(0,5), ha='center', va='bottom', fontsize=8, color='blue')

# Add data labels for Female Voters
for i, txt in enumerate(df2['Female Voters']):
    ax.annotate(f'{txt} ({df2["Female Percentage"][i]:.2f}%)', (df2['Year'][i], df2['Female Voters'][i]), textcoords="offset points", xytext=(0,5), ha='center', va='bottom', fontsize=8, color='#AA336A')

# Beautify legend
ax.legend()
# Save the plot before showing it
plt.savefig('E:/Project/Resolvent Projects/C.G.Election Analysis/Analysis/Number of Male and Female Voters.png')

# Show the plot
plt.show()
```

```python
# Calculate percentages for Male and Female Vote Casted
df2['Male Vote Casted Percentage'] = (df2['Male Vote Casted'] / df2['No. of Voters']) * 100
df2['Female Vote Casted Percentage'] = (df2['Female Vote Casted'] / df2['No. of Voters']) * 100

# Increase figure size
fig, ax = plt.subplots(figsize=(14, 8))

# Plotting lines for Male Voters, Female Voters, Male Vote Casted, and Female Vote Casted
ax.plot(df2['Year'], df2['Male Voters'], marker='o', linestyle='-', color='blue', label='Male Voters')
ax.plot(df2['Year'], df2['Female Voters'], marker='o', linestyle='-', color='#AA336A', label='Female Voters')
ax.plot(df2['Year'], df2['Male Vote Casted'], marker='o', linestyle='--', color='blue', label='Male Vote Casted')
ax.plot(df2['Year'], df2['Female Vote Casted'], marker='o', linestyle='--', color='#AA336A', label='Female Vote Casted')

# Improve labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Count')
ax.set_title('Year-wise Male Voters, Female Voters, Male Vote Casted, and Female Vote Casted with Percentages')

# Add data labels for Male Voters
for i, txt in enumerate(df2['Male Voters']):
    ax.annotate(f'{txt} ({df2["Male Percentage"][i]:.2f}%)', (df2['Year'][i], df2['Male Voters'][i]), textcoords="offset points", xytext=(0,5), ha='center', va='bottom', fontsize=8, color='blue')

# Add data labels for Female Voters
for i, txt in enumerate(df2['Female Voters']):
    ax.annotate(f'{txt} ({df2["Female Percentage"][i]:.2f}%)', (df2['Year'][i], df2['Female Voters'][i]), textcoords="offset points", xytext=(0,5), ha='center', va='bottom', fontsize=8, color='#AA336A')

# Add data labels for Male Vote Casted
for i, (txt, percentage) in enumerate(zip(df2['Male Vote Casted'], df2['Male Vote Casted Percentage'])):
    ax.annotate(f'{txt:.0f} ({percentage:.2f}%)', (df2['Year'][i], df2['Male Vote Casted'][i]), textcoords="offset points", xytext=(0,5), ha='center', va='bottom', fontsize=8, color='blue')

# Add data labels for Female Vote Casted
for i, (txt, percentage) in enumerate(zip(df2['Female Vote Casted'], df2['Female Vote Casted Percentage'])):
    ax.annotate(f'{txt:.0f} ({percentage:.2f}%)', (df2['Year'][i], df2['Female Vote Casted'][i]), textcoords="offset points", xytext=(0,5), ha='center', va='bottom', fontsize=8, color='#AA336A')

# Beautify legend
ax.legend()

# Save the plot before showing it
plt.savefig('E:/Project/Resolvent Projects/C.G.Election Analysis/Analysis/Number of Male and Female Vote Casted.png')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming df2 is your DataFrame

# Calculate the percentage of winner and loser votes relative to total votes cast
df2['Winner Vote Percentage'] = (df2['Winner Vote'] / df2['Vote Casted']) * 100
df2['Loser Vote Percentage'] = (df2['Looser Vote'] / df2['Vote Casted']) * 100

# Create a figure and axis with proper spacing
fig, ax = plt.subplots(figsize=(16, 8))
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.3)

# Set the width of the bars
bar_width = 0.35

# Plotting bar graphs for Winner and Looser Parties every year with increased linewidth
ax.bar(df2['Year'] - bar_width / 2, df2['Winner Vote Percentage'], label='Winner Vote Percentage', width=bar_width, linewidth=2)
ax.bar(df2['Year'] + bar_width / 2, df2['Loser Vote Percentage'], label='Loser Vote Percentage', width=bar_width, linewidth=2)

# Add data labels for Winner Vote Percentage
for i, (percentage, party, candidate) in enumerate(zip(df2['Winner Vote Percentage'], df2['Winner Party'], df2['Winner Candidate '])):
    height = df2['Winner Vote Percentage'][i]
    ax.annotate(f'{party}\n{candidate}\n{percentage:.2f}%', 
                (df2['Year'][i] - bar_width / 2, height), 
                textcoords="offset points", xytext=(0, 10 if height < 10 else -10), 
                ha='center', va='bottom', fontsize=8, fontweight='bold', color='blue' if height < 10 else 'black')

# Add data labels for Loser Vote Percentage
for i, (percentage, party, candidate) in enumerate(zip(df2['Loser Vote Percentage'], df2['Looser Party'], df2['Looser Candidate'])):
    height = df2['Loser Vote Percentage'][i]
    ax.annotate(f'{party}\n{candidate}\n{percentage:.2f}%', 
                (df2['Year'][i] + bar_width / 2, height), 
                textcoords="offset points", xytext=(0, 10 if height < 10 else -10), 
                ha='center', va='bottom', fontsize=8, fontweight='bold', color='orange' if height < 10 else 'black')

# Improve labels and title
ax.set_xlabel('Year', fontsize=12, fontweight='bold')
ax.set_ylabel('Vote Percentage', fontsize=12, fontweight='bold')
ax.set_title('Winner and Loser Vote Percentages Every Year', fontsize=14, fontweight='bold')

# Remove the legend
ax.legend().set_visible(False)

# Save the plot before showing it
plt.savefig('E:/Project/Resolvent Projects/C.G.Election Analysis/Analysis/Winner_Loser_Vote_Percentages.png')

# Show the plot
plt.show()
```
