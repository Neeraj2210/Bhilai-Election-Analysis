```python
## Data Analysis Phase
## MAin aim is to understand more about the data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
## Display all the columns of the dataframe
import pyodbc
pd.pandas.set_option('display.max_columns',None)
```

```python
df=pd.read_excel("E:/Project/Resolvent Projects/C.G.Election Analysis/Analysis/Stage5.xlsx")
```

```python
df.head()
```

```python
# Convert numeric columns to strings
df['LS'] = df['LS'].astype(str)
df['VS'] = df['VS'].astype(str)
df['District'] = df['District'].astype(str)
df['PinCode'] = df['PinCode'].astype(str)
df['VillageCity'] = df['VillageCity'].astype(str)
df['Anubhag'] = df['Anubhag'].astype(str)

# Concatenate multiple columns into 'Address' column
df['Address'] = (
    df['LS'] + ', ' +
    df['VS'] + ', ' +
    df['District'] + ', ' +
    df['PinCode'] + ', ' +
    df['VillageCity'] + ', ' +
    df['Anubhag'] + ', ' +
    df['Address']
)
```

```python
df.head()
```

```python
import re

# Define a function to clean text data
def clean_text(text):
    if isinstance(text, str):
        # Remove special characters, 1 or 2 letter words, and leading/trailing whitespaces
        cleaned_text = re.sub(r'[^a-zA-Z ]|(\b\w{1,2}\b)', '', text).strip()
        return cleaned_text
    else:
        return text

# Apply the clean_text function to specified columns
columns_to_clean = ['Age']
for col in columns_to_clean:
    df[col] = df[col].apply(clean_text)
```

```python
df.info()
```

```python
df['Age'].fillna(0, inplace=True)  # Fill NaN values with 0 or any other default value
df['Age'] = df['Age'].astype(int)
```

```python
df.head()
```

```python
df.drop(columns=['Name','Sno','Relative_Name','Word_Count','NumDigits'], inplace=True)
```
