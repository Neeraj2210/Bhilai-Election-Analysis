```python
## Here we will check the percentage of nan values present in each feature
## 1 -step make the list of features which has missing values
df_features_with_na=[features for features in df.columns if df[features].isnull().sum()>1]
## 2- step print the feature name and the percentage of missing values

for feature in df_features_with_na:
    print(feature, df[feature].isnull().sum(),  'missing values')
    print(feature, np.round(df[feature].isnull().mean(), 2),  ' % missing values')
```

```python
df_categorical_features=[feature for feature in df.columns if df[feature].dtypes=='O']

for feature in df_categorical_features:
    print('The feature is {} and number of categories are {}'.format(feature,len(df[feature].unique())))
```

```python
#Gender
df['Gender'].unique()
```

```python
# Replace variations of 'Female,' 'Woman,' 'Women,' 'woman,' 'Female AA,' and 'Female A' with 'female' (case-insensitive)
df['Gender'] = df['Gender'].str.replace(r'(?i)^(female|Woman|Women|Female | Female|Female)$', 'Female')
df['Gender'] = df['Gender'].str.replace(r'(?i)^(male|Male GA|Male|male aa|male|Men|male high| Male)$', 'Male')

# Display the updated DataFrame
df['Gender'].unique()
# Delete rows where 'Gender' is non-null
df = df[df['Gender'].notna()]
```

```python
df['Gender'].unique()
```

```python
#Surname

df['Surname'].unique()
```

```python
# Count the occurrences of each unique surname
surname_counts = df['Surname'].value_counts().reset_index()

# Rename the columns for clarity
surname_counts.columns = ['Surname', 'Count']

# Determine the top 6 surnames and their counts
top_surnames = surname_counts.head(15)

# Create a mask to identify rows with surnames not in the top 6
mask = ~df['Surname'].isin(top_surnames['Surname'])

# Replace those surnames with 'Others'
df.loc[mask, 'Surname'] = 'Others'

# Display the DataFrame with marked 'Others' for remaining surnames

surname_counts.columns = ['Surname', 'Count']

# Display the DataFrame with surname counts
print(surname_counts)
```

```python
df['Surname'].value_counts()
```

```python
# Populate the 'Name' and 'Relation' columns based on 'Father Name' and 'Husband Name'
df['Relative_Name'] = df['Father Name'].fillna('') + ' / ' + df['Husband Name'].fillna('')
df['Relation'] = df['Father Name'].apply(lambda x: 'Father' if pd.notna(x) else '') + df['Husband Name'].apply(lambda x: 'Husband' if pd.notna(x) else '')
df['Relative_Name'] = df['Relative_Name'].str.replace(' / ', '')
df.drop(columns=['Father Name','Husband Name'], inplace=True)
df.head()
```

```python
# Age

df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
# Define the age groups and corresponding labels
age_bins = [18, 25, 32, 42, 52, 60, 70, float('inf')]
age_labels = ['18-25', '26-32', '33-42', '43-52', '53-60', '61-70', '70+']

# Use pd.cut() to categorize ages into groups and add a new 'Age Group' column
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
df.head()
```

```python
#Surname
df['Surname'] = df['Candidate Name'].str.split().str[-1]
df.head()
```

```python
#Address
# Consolidate address columns into one 'Address' column
# Convert 'Pin Code' column to string
# Convert 'Pin Code' column to string
df['Pin Code'] = df['Pin Code'].astype(str)
df['Address'] = df['Address'].astype(str)
# Concatenate multiple columns into 'Address' column
df['Address'] = (
    df['Address'] + ', ' +
    df['Anubhag Name'] + ', ' +
    df['Village Name'] + ', ' +
    df['Post Office'] + ', ' +
    df['Panchayat Name'] + ', ' +
    df['District Name'] + ', ' +
    df['Vidhansabha Name'] + ', ' +
    df['Lokshabha Name'] +
    ', Pin code=' + df['Pin Code']
)

# Display the resulting DataFrame
df.head()
```

```python
df['No. of Family member_Group'] = df.groupby('Address').ngroup()
df.head()
```

```python
print(df.groupby('Address'))
```

```python
# Custom mapping of unique counts to categories
category_mapping = {
    1: 'first',
    2: 'second',
    3: 'Third',
    4: 'Fourth',
    5: 'Fifth',
    6: 'Sixth',
    7: 'Seventh',
    8: 'Eighth',
    9: 'Ninth',
    10: 'Tenth',
    11: 'Eleventh',
    12: 'Twelfth',
    13: 'Thirteenth',
    14: 'Fourteenth',
    15: 'Fifteenth',
    16: 'Sixteenth',
    17: 'Seventeenth',
    18: 'Eighteenth',
    19: 'Nineteenth',
    20: 'Twentieth',
    21: 'Twenty-First',
    22: 'Twenty-Second',
    23: 'Twenty-Third'
    # Add more categories if needed
}

# Group the DataFrame by 'Address' and 'Candidate Name' and count the number of unique members in each group
unique_address_candidate_counts = df.groupby(['Address', 'Candidate Name']).size().unique()

# Create a pandas Series from the NumPy array and use map with the custom mapping
category_series = pd.Series(unique_address_candidate_counts)
df['Category'] = category_series.map(category_mapping)

# Display the DataFrame with the 'Category' column
df.head()
```

```python
# Group the DataFrame by 'Address' and count the number of people in each group
grouped = df.groupby('Address').size().reset_index(name='Group Size')

# Merge the group size information back into the original DataFrame
df = df.merge(grouped, on='Address')

# Display the DataFrame with 'Group Size' column
df.head()
```

## District wise Gender Distribution

```python
df.dropna(subset=['Relation'], inplace=True)
```

```python
# Set a custom color palette
colors = ["#85c1e9", "#f1948a"]
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=colors)

# Create a count plot for the 'Gender' column against 'District'
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=df, x='District Name', hue='Gender')

# Customize the plot
plt.xlabel('District Name', fontsize=12, fontweight='bold')
plt.ylabel('Count', fontsize=12, fontweight='bold')
plt.title('Gender Count by District', fontsize=14, fontweight='bold')
plt.legend(title='Gender', loc='upper right')

# Add percentage labels on top of the bars
total_height = len(df)
for p in ax.patches:
    height = p.get_height()
    percentage = (height / total_height) * 100
    ax.annotate(f'{percentage:.1f}%', (p.get_x() + p.get_width() / 2., height), ha='center', va='bottom', fontsize=10, color='black', xytext=(0, 10), textcoords='offset points')

# Customize legend labels
legend_labels = ax.get_legend().get_texts()
for label, gender in zip(legend_labels, ['Male', 'Female']):
    label.set_text(f'{gender} ({df[df["Gender"] == gender].shape[0]})')

# Show the plot
plt.tight_layout()
plt.show()
```

## Anubhag distribution in village

## गाँव में अनुभग वितरण

```python
# Group the data by 'Village Name' and calculate the percentage distribution of 'Anubhag Name' within each village
village_groups = df.groupby('Village Name')['Anubhag Name'].value_counts(normalize=True) * 100

# Get a list of unique village names
unique_villages = df['Village Name'].unique()

# Set up the subplots
fig, axes = plt.subplots(1, len(unique_villages), figsize=(15, 5), sharey=True)

# Colors for the pie chart slices
colors = ['#85c1e9', '#f1948a', '#99ff99', '#ffb3e6', '#c2c2f0', '#f0e68c', '#ff6666', '#c2f0c2', '#ffdb4d', '#ff9999',
          '#66b3ff', '#c2f0f0', '#99ff99', '#ffcc99', '#b3b3cc', '#99cc99', '#ffdb58', '#c2b3f0']

# Iterate through each unique village and create a pie chart
for i, village in enumerate(unique_villages):
    # Get the percentage distribution for the current village
    data = village_groups[village].sort_index()

    # Create a pie chart for the current village
    ax = axes[i]
    ax.pie(data, labels=data.index, colors=colors, autopct='%1.1f%%', startangle=140)
    ax.set_title(f'Village: {village}', fontsize=12, fontweight='bold')

# Add a common title for all subplots
fig.suptitle('Percentage Distribution of Anubhag Name by Village', fontsize=14, fontweight='bold')

# Adjust spacing between subplots
plt.tight_layout()

# Display the subplots
plt.show()
```
